```python
import pandas as pd
from transformers import pipeline

# 1. Load your dataset (only first 23000 records)
df = pd.read_csv("ner.csv").head(23000).reset_index(drop=True)

# 2. Initialize the Hugging Face NER pipeline
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", aggregation_strategy="simple")

# 3. Prepare a list to store results
results = []

# 4. Loop through each text row in your dataset
for idx, text in enumerate(df['text']):
    entities = ner_pipeline(text)
    # Format entities as a string for saving, e.g. "word:entity (score)"
    entity_strs = [f"{ent['word']}:{ent['entity_group']}({ent['score']:.2f})" for ent in entities]
    entities_joined = "; ".join(entity_strs)

    # Append original text + entities recognized
    results.append({
        "text": text,
        "entities": entities_joined
    })

    # Optional: print progress every 100 rows
    if idx % 100 == 0:
        print(f"Processed {idx} texts")

# 5. Save results as a new CSV file
output_df = pd.DataFrame(results)
output_df.to_csv("ner_predictions_output.csv", index=False)

print("NER predictions saved to ner_predictions_output.csv")
```
